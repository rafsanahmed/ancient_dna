### Split raw data and decision tree

```python
df_main = df_main[numeric_cols]
x = df_main.drop("DateBP", axis=1).values
y = df_main["DateBP"].values

print(x.shape)
print(y.shape)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, test_size=0.2, random_state=123)
print(x_train.shape, x_test.shape, y_train.shape,y_test.shape)
```

```python
rf = RandomForestRegressor(n_estimators=100,random_state=123)
rf.fit(x_train, y_train)
```

```python
# check importance of feature
for i, column in enumerate(df_main.drop("DateBP", axis=1)):
    print("importance of {}: {}".format(column,rf.feature_importances_[i]))
```

```python
# Train accuracy 
print("Train accuracy: {}".format(rf.score(x_train,y_train)))

# Test accuracy 
print("Test accuracy: {}".format(rf.score(x_test,y_test)))
```

```python
y_pred = rf.predict(x_test)
print(y_pred[:10])
print("MSE: {}".format(mean_squared_error(y_test, y_pred)))
print("R^2: {}".format(r2_score(y_test, y_pred)))
```

### Fine tuning RF

```python
from itertools import product
list_df_finetune = []
n_estimators = [10,100,1000]
max_features = [1, "sqrt", "log2"]
max_depth = [None, 2,3,4,5,10]

for n,f,d in product(n_estimators,max_features, max_depth):
    rf_ = RandomForestRegressor(n_estimators=n, max_features=f,
                               max_depth=d,n_jobs=2, random_state=123)
    rf_.fit(x_train, y_train)
    pred_test = rf_.predict(X=x_test)
    list_df_finetune.append([n,f,d,mean_squared_error(y_test, pred_test),
                                                          r2_score(y_test, pred_test)])

df_rf_finetune = pd.DataFrame(list_df_finetune, columns=["num_estimators", "max_features", "max_depth", "MSE", "R^2"])
df_rf_finetune
```

```python
# min MSE
df_rf_finetune[df_rf_finetune["MSE"]==df_rf_finetune["MSE"].min()]
```

```python
# Max R^2
df_rf_finetune[df_rf_finetune["R^2"]==df_rf_finetune["R^2"].max()]
```

```python
classifier = xgboost.XGBRegressor(tree_method="gpu_hist")

params = {
    "learning_rate":[0.05,0.1,0.15,0.2,0.25,0.3],
    "max_depth":[2,3,4,5,6,8,10,12,15],
    "max_child_weight":[1,3,5,7],
    "gamma":[0.0,0.1,0.2,0.3,0.4],
    "colsample_bytree":[0.3,0.4,0.5,0.7]}

clf = RandomizedSearchCV(classifier, param_distributions=params, n_iter=5,scoring='roc_auc', cv=5,verbose=3)

clf.fit(x_train,y_train)
```
